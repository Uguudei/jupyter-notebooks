```python
import datetime
import pandas as pd
import requests
from pprint import pprint
from pandas.io.json import json_normalize
```

```python
api_key = 'api_key'
api_url = 'http://api.openweathermap.org/data/2.5/'
unit = 'metric'
city_id = '2028462'
lat = 47.90
lon = 47.91
```

#### Weather

```python
# Current weather
"""https://openweathermap.org/current"""
api_type = 'weather'
payload = {'id': city_id, 'units': unit, 'appid': api_key}
r = requests.get(api_url+api_type, params=payload)
print(r.url)
r.json()
```

```python
# Forecast
"""https://openweathermap.org/forecast5"""
api_type = 'forecast'
payload = {'id': city_id, 'units': unit, 'appid': api_key, 'cnt': 5}
r = requests.get(api_url+api_type, params=payload)
print(r.url)
forecast_dict = r.json()
pprint(r.json())
```

```python
df = json_normalize(forecast_dict)
forecast_dict.keys()
pd.DataFrame(forecast_dict['list'])
```

#### UV

```python
# Current UV index
"""https://openweathermap.org/api/uvi"""
api_type = 'uvi'
payload = {'appid': api_key, 'lat': lat, 'lon': lon}
r = requests.get(api_url+api_type, params=payload)
print(r.url)
pprint(r.json())
```

```python
# UV index forecast
"""https://openweathermap.org/api/uvi"""
api_type = 'uvi/forecast'
payload = {'appid': api_key, 'lat': lat, 'lon': lon}
r = requests.get(api_url+api_type, params=payload)
print(r.url)
pprint(r.json())
```

```python
# UV index historical
"""https://openweathermap.org/api/uvi"""
api_type = 'uvi/history'
payload = {'appid': api_key, 'lat': lat, 'lon': lon, 'start': 1498049953, 'end': 1498481991}
r = requests.get(api_url+api_type, params=payload)
print(r.url)
pprint(r.json())
```

#### One Call

```python
# Current weather
"""https://openweathermap.org/api/one-call-api"""
api_type = 'onecall'
# {'id': city_id, 'units': unit, 'appid': api_key}
payload = {'appid': api_key, 'lat': lat, 'lon': lon, 'units': unit, 
           'exclude': 'minutely, hourly, daily'}
r = requests.get(api_url+api_type, params=payload)
print(r.url)
pprint(r.json())
```

```python
# Hourly weather
"""https://openweathermap.org/api/one-call-api"""
api_type = 'onecall'
payload = {'appid': api_key, 'lat': lat, 'lon': lon, 'units': unit,
           'exclude': 'current, minutely, daily'}
r = requests.get(api_url+api_type, params=payload)
print(r.url)
pprint(r.json())
```

```python
# Daily weather
"""https://openweathermap.org/api/one-call-api"""
api_type = 'onecall'
payload = {'appid': api_key, 'lat': lat, 'lon': lon, 
           'exclude': 'current, minutely, hourly'}
r = requests.get(api_url+api_type, params=payload)
print(r.url)
pprint(r.json())
```
