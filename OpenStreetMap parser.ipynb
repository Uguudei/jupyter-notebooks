```
import pandas as pd
import os
from google.colab import drive, files
drive.mount('/content/drive')
```

```
# Import data
maposm = os.path.join('data', '/content/drive/My Drive/Colab Notebooks/Geospacial Analysis/mongolia_osm_raw_data')
```

```
import xml.etree.cElementTree as et
parsedXML = et.parse(maposm)
root = parsedXML.getroot()
```

```
# %%timeit
key_list = ['name', 'name:en', 'name:mn', 'amenity', 'shop', 'tourism', 'leisure', 'office', 'craft', 'operator']
node_list = []
for node in root:
    node_temp = {}
    node_add = False
    node_children = node.getchildren()
    if node_children != [] and node.attrib.get('lat') != None and node.attrib.get('lon') != None:
        for child in node_children:
            k = child.attrib.get('k')
            v = child.attrib.get('v')
            if k in key_list:
                node_temp[k] = v
                node_add = True
        if node_add == True:
            # node_temp['tag'] = node.tag
            node_temp['id'] = node.attrib.get('id')
            node_temp['lat'] = node.attrib.get('lat')
            node_temp['lon'] = node.attrib.get('lon')
            node_list.append(node_temp)
```

```
df_node = pd.DataFrame(node_list)
print(df_node.head(3))
```
